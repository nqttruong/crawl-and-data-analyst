```python
import pandas as pd
import os 
import numpy as np
```

```python
path = r"D:\CNTT\py_for_da\crawl_data_analyst\tiki\dienthoai_maytinhbang\maytinhbang"
os.listdir(path)
```

```python
frames = []
for file in os.listdir(path):
    if file.endswith('.csv'):
        filepath = path + '\\' + file
        df1 = pd.read_csv(filepath)
        frames.append(df1)
        result = pd.concat(frames)

df_all = result
df_all.to_csv("tk_maytinhbang_all.csv", index="False")
```

1. Xử lý dữ liệu

```python
df_all["category"] = "maytinhbang"
df_all = df_all.reset_index(drop=True)
df_all = df_all.drop('index', axis=1)
```

```python
df_all
```

```python
df_all["price"] = df_all["price"].replace('[\₫,]', '', regex=True).replace('\.', '', regex=True).astype(int)
print(df_all["price"])
```

```python
def convert_to_number(value):
    if isinstance(value, str):
        if 'k+' in value:
            return float(value.replace('k+', '')) * 1000
        elif 'k' in value:
            return float(value.replace('k', '')) * 1000
        else:
            return float(value) if value else None
    else:
        return value
```

```python
# Áp dụng hàm convert_to_number cho cột 'Gia'
df_all["sales"] = df_all["sales"].apply(lambda x: np.nan if pd.isna(x) else x.split('Đã bán')[1].strip())
df_all["sales"] = df_all["sales"].apply(convert_to_number)
df_all["sales"]
```

```python
df_all
```

```python
df_all["revenue"] = df_all["price"] * df_all["sales"]
df_all["revenue"]
```

```python
total = df_all["revenue"].sum()
```

```python
print("Tổng doanh số máy tính bảng: ", total)

```
